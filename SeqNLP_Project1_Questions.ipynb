```python
from keras.datasets import imdb

vocab_size = 10000 #vocab size
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size) # vocab_size is no.of words to consider from the dataset, ordering based on frequency.
```

```python
# get the word-index 
word2id = imdb.get_word_index()

# create a dict of work-id and word
id2word = {i: word for word, i in word2id.items()}
```

## Pad Sequences

```python
from keras.preprocessing.sequence import pad_sequences
maxlen = 300  #number of word used from each review
```

```python
#make all sequences of the same length
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test =  pad_sequences(x_test, maxlen=maxlen)
```

## Build a Sequential Model using Keras for the Sentiment Classification task (20 points) 

```python
from keras import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout
embedding_size=50

model=Sequential()

model.add(Embedding(vocab_size, #Vocabulory size
                    embedding_size, #Embedding size
                    input_length=maxlen)) #Number of words in each review

model.add(LSTM(256, #RNN State - size of cell state and hidden state
            dropout=0.2, #Dropout before feeding the data to LSTM layer
            recurrent_dropout=0.2)) #Dropout applied to the output of LSTM layer

model.add(Dense(1, activation='sigmoid'))
print(model.summary())
```

```python
#compile the model 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

## Report the Accuracy of the model (10 points)  

```python
# Run the model 
model.fit(x_train,y_train,
          epochs=20,
          batch_size=32,          
          validation_data=(x_test, y_test))
```

```python
scores = model.evaluate(x_test, y_test, verbose=0)
```

```python
print("Test Loss And Accuracy:", scores)
```

## Retrieve the output of each layer in Keras for a given single test sample from the trained  model you built (10 points) 


```python
# model layers
model.layers[0], model.layers[1], model.layers[2]
```

### Embedded Layer

```python
from keras import backend

# Function for embedded-layer 
get_embed_out = backend.function(
    [model.layers[0].input], #Embedding Layer Input = Batch_Size * Length of each review
    [model.layers[0].output]) #Embedding Layer Output - [Batch_Size , Review Length , Embedding_Size]

#Embedded layers
model.layers[0].input, [model.layers[0].output]
```

```python
# get the embedded-layer output for a test sample x_test[0:1]
embed_layer_output = get_embed_out([x_test[0:1,:]])
print(type(embed_layer_output), len(embed_layer_output), embed_layer_output[0].shape)
```

```python
embed_layer_output[0]
```

### LSTM Layer

```python
# Function for LSTM Layer
get_LSTM_layer_out = backend.function(
    [model.layers[1].input], 
    [model.layers[1].output]) 

# get the LSTM-layer output for a test sample above
lstm_layer_output = get_LSTM_layer_out(embed_layer_output)
```

```python
print(type(lstm_layer_output), len(lstm_layer_output), lstm_layer_output[0].shape)
```

###Dense Layer


```python
# Function for Dense Layer
get_dense_layer_out = backend.function(
    [model.layers[2].input], 
    [model.layers[2].output]) 

# get the Dense-Layer output for a test sample above
dense_layer_output = get_dense_layer_out(lstm_layer_output)
```
