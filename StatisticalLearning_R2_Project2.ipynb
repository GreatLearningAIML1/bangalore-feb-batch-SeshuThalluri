```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, wilcoxon
```

```python
data_imp = pd.read_csv("data.csv")
data_imp['Old Scheme (in thousands)'] = data_imp['Old Scheme (in thousands)'] *1000
data_imp['New Scheme (in thousands)'] = data_imp['New Scheme (in thousands)'] *1000
data_imp.describe()
```

### 1.Find the mean of old scheme and new scheme column

```python
old_scheme_mean = data_imp['Old Scheme (in thousands)'].mean()
new_scheme_mean = data_imp['New Scheme (in thousands)'].mean()

print('Old-Scheme Mean ={} , New-Scheme Mean = {}\n'.format(old_scheme_mean, new_scheme_mean))
```

### 2.Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs?

```python
# Seperating data into 2 groups
old_scheme = np.array(data_imp['Old Scheme (`in thousands)'])
new_scheme = np.array(data_imp['New Scheme (in thousands)'])
```

```python
(new_scheme - old_scheme).mean()
(new_scheme - old_scheme).std()
```

```python
# paired t-test: doing two measurments on the same experimental unit
# Null Hypothesis: Ud = U2-U1 = 0
# Alternative Hypothesis: Ud = U2-U1 # 0
t_statistic, p_value = ttest_1samp(new_scheme - old_scheme, 0)
print('t_statistic={}, p_value={}\n'.format(t_statistic, p_value))
```

### 3.What conclusion does the test (p-value) lead to?

```python
# p_value is > 0.05, Do Not Reject Null Hypothesis
print ("paired t-test p-value=", p_value)
print ("Do Not Reject Null Hypothesis (Ud = U2-U1 = 0), There is no significant raise in new scheme")
```

#### What is the p- value of the hypothesis test if we test for a difference of $5000?

```python
# Null Hypothesis: Ud = U2-U1 <= 5000
# Alternative Hypothesis: Ud > 5000 
t_statistic, p_value = ttest_1samp(new_scheme - old_scheme, 5000)
print('t_statistic={}, p_value={}\n'.format(t_statistic, p_value))
```

```python
# T-Test Critical-Values for alpha=0.05 and Sample-Size=30 is -2.045 to 2.045. As we are only looking Upper-Tail T-test, 
# so the Critical-Value to be considered is 2.045

# Tstat is '-0.389'. Tstat (-0.389) < Critical-value (2.045). Hence, there is no sufficient evidence to say that Ud > 5000. 
# Do not Reject the Null Hypothesis (Ud <= 5000). 

#p_value needs to be calculated only for Upper-Tail T-Test. So, p_value is (1 - (0.70/2)) = 0.65
p_value_upper_tail_t_test = (1 - (0.70/2))
print ("p_value for Upper-Tail T-Test=", p_value_upper_tail_t_test)

# p_value_upper_tail_t_test of 0.65 > 0.05. So, Do Not Reject the Nul Hypothesis. 
```

#### Power of the test

```python
# Calculating Power of Test
# Compute the difference in Means between 2 sample means and divide by pooled Standard Deviation 
# number of Observations/tuples
# Set the alpha value to 0.05 and alternative values 'two-sided' , 'larger' , 'smaller'
eff_val = (np.mean(new_scheme-old_scheme) - 5000) / np.std(new_scheme - old_scheme)
eff_val
```

```python
from statsmodels.stats.power import ttest_power
print ("Power of Test =", ttest_power(eff_val, nobs=len(new_scheme), alpha=0.05, alternative='larger'))
```
